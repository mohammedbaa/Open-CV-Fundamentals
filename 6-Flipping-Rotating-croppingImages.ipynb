## Flipping-Rotating-cropping Images

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image=cv2.imread("images/giraffe-Kenya.png")
print(image)

plt.imshow(image)
```

```python
img_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

## Flip 


```python
## 1 --> Horizantal 
## 0 --> Virtical
## -1 --> Both Axis  
flip_horizontel=cv2.flip(image,1)
flip_Virtical=cv2.flip(image,0)
flip_Both=cv2.flip(image,-1)


titles=["Orginal Images" , "Horizantal Flip","Virtical Flip","Both"]
images=[img_rgb ,cv2.cvtColor(flip_horizontel,cv2.COLOR_BGR2RGB),cv2.cvtColor(flip_Virtical,cv2.COLOR_BGR2RGB),cv2.cvtColor(flip_Both,cv2.COLOR_BGR2RGB)]


plt.figure(figsize=(12,8))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.show()
```

## Rotate 

```python
rotate_90_clock=cv2.rotate(img_rgb,cv2.ROTATE_90_CLOCKWISE)
rotate_90_COUNTERclock=cv2.rotate(img_rgb,cv2.ROTATE_90_COUNTERCLOCKWISE)
rotate_180=cv2.rotate(img_rgb,cv2.ROTATE_180)


titles=["Orginal Images" , "90_clock","90_COUNTERclock","rotate_180"]
images=[img_rgb ,rotate_90_clock,
        rotate_90_COUNTERclock,
        rotate_180]



plt.figure(figsize=(12,8))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.show()
```

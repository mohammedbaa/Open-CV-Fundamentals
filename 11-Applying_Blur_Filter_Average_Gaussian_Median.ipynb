## Apply Blur Filter (Average , Gaussin , Median)

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
image=cv2.imread('images/giraffe-Kenya.png')
```

```python
img_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

image_gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
kernal_s=5
Average_blur=cv2.blur(image_gray,(kernal_s,kernal_s))
plt.imshow(Average_blur,cmap='gray')
```

```python
kernal_s=31
Gaussian_blur=cv2.GaussianBlur(image_gray,(kernal_s,kernal_s),0)
plt.imshow(Gaussian_blur,cmap='gray')
```

```python
median_blur=cv2.medianBlur(image_gray,13)
plt.imshow(median_blur,cmap='gray')
```

```python
plt.figure(figsize=(14,12))
plt.subplot(2,2,1)
plt.title("Original grayscale")
plt.imshow(image_gray,cmap='grey')

plt.subplot(2,2,2)
plt.title("Average_blur grayscale")
plt.imshow(Average_blur,cmap='grey')


plt.subplot(2,2,3)
plt.title("Gaussian_blur grayscale")
plt.imshow(Gaussian_blur,cmap='grey')


plt.subplot(2,2,4)
plt.title("median_blur grayscale")
plt.imshow(median_blur,cmap='grey')
```

## Image Filter 

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
image=cv2.imread('images/filter.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(image,cmap='gray')
```

```python
horizantal=np.array([[-1,-1,-1],[0,0,0],[1,1,1]])
horizantal_filter_image=cv2.filter2D(image,-1,horizantal)

plt.imshow(horizantal_filter_image,cmap='gray')
```

```python
horizantal=np.array([[-1,-1,-1],[5,5,5],[1,1,1]])
horizantal_filter_image=cv2.filter2D(image,-1,horizantal)

plt.imshow(horizantal_filter_image,cmap='gray')
```

```python
horizantal=np.array([[-1,-1,-1],[-5,-5,-5],[1,1,1]])
horizantal_filter_image=cv2.filter2D(image,-1,horizantal)

plt.imshow(horizantal_filter_image,cmap='gray')
```

```python
horizantal=np.array([[-1,-1,-1],[-10,-10,-10],[1,1,1]])
horizantal_filter_image=cv2.filter2D(image,-1,horizantal)

plt.imshow(horizantal_filter_image,cmap='gray')
```

```python
x=-15
horizantal=np.array([[-1,-1,-1],[x,x,x],[1,1,1]])
horizantal_filter_image=cv2.filter2D(image,-1,horizantal)

plt.imshow(horizantal_filter_image,cmap='gray')
```

```python
x=-15
Virtical=np.array([[-1,x,1],
                   [-1,x,1],
                   [-1,x,1]])
Virtical_filter_image=cv2.filter2D(image,-1,Virtical)

plt.imshow(Virtical_filter_image,cmap='gray')
```

```python
x=-15
diognal=np.array(
    [[x,-1,-1]
    ,[1,x,-1],
     [1,1,x]])
diognal_filter_image=cv2.filter2D(image,-1,diognal)

plt.imshow(diognal_filter_image,cmap='gray')
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.title("Original Image")
plt.imshow(image,cmap='gray')


plt.subplot(2,2,2)
plt.title("Horizantal Image")
plt.imshow(horizantal_filter_image,cmap='gray')



plt.subplot(2,2,3)
plt.title("Vertical Image")
plt.imshow(Virtical_filter_image,cmap='gray')




plt.subplot(2,2,4)
plt.title("diognal Image")
plt.imshow(diognal_filter_image,cmap='gray')


plt.show()
```

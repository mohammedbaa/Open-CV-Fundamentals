## Edge Detection 

```python
import cv2 
import numpy 
import matplotlib.pyplot as plt
```

```python
image=cv2.imread('images/giraffe-Kenya.png',cv2.IMREAD_GRAYSCALE)


### Sobel Edge Detector 

sobel_x=cv2.Sobel(image,cv2.CV_64F,1,0,ksize=3)
sobel_y=cv2.Sobel(image,cv2.CV_64F,0,1,ksize=3)
sobel_combiend=cv2.magnitude(sobel_x,sobel_y)


### Laplacian
Laplacian = cv2.Laplacian(image,cv2.CV_64F , ksize=3)
Laplacian_abs=cv2.convertScaleAbs(Laplacian)


## Canny 
canny_edge=cv2.Canny(image,100,100)


```

```python
title=['Original Image' ,'Sobel' , 'Laplacian_All' , 'Laplacian_abs' , 'canny_edge' ]
images=[image ,sobel_combiend , Laplacian , Laplacian_abs , canny_edge ]
plt.figure(figsize=(10,8))
for i in range(len(title)):
    plt.subplot(3,2,i+1)
    plt.imshow(images[i],cmap='gray')
    plt.title(title[i])
    plt.axis('off')
plt.show()


```

## CLAHE

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
image=cv2.imread('images/giraffe-Kenya.png')
```

```python
image_gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

equlized_image=cv2.equalizeHist(image_gray)

clahe=cv2.createCLAHE(clipLimit=2.0 , tileGridSize=(4,4))  
clahe_equalized_image=clahe.apply(image_gray)

plt.figure(figsize=(12,8))

plt.subplot(1,3,1)
plt.imshow(image_gray,cmap='gray')
plt.title("original image")


plt.subplot(1,3,2)
plt.imshow(equlized_image,cmap='gray')
plt.title("Equalized image")


plt.subplot(1,3,3)
plt.imshow(clahe_equalized_image,cmap='gray')
plt.title("Gray Equalized image")


plt.show()
```

```python
Image_yuv=cv2.cvtColor(image,cv2.COLOR_BGR2YUV)

Image_yuv[:,:,0]=clahe.apply(Image_yuv[:,:,0]) 


Image_yuv_BGR=cv2.cvtColor(Image_yuv,cv2.COLOR_YUV2BGR)
Image_yuv_RGB=cv2.cvtColor(Image_yuv_BGR,cv2.COLOR_BGR2RGB)

image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)




plt.figure(figsize=(12,8))

plt.subplot(1,3,1)
plt.imshow(image_rgb)
plt.title("original image")


plt.subplot(1,3,2)
plt.imshow(Image_yuv_RGB)
plt.title(" Equalized RGB image")


plt.show()
```

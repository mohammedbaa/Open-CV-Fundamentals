## Countours

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
image=cv2.imread('images/shapes.jpg')
plt.imshow(image)
```

```python
image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image_gray=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(image_rgb)
```

```python
gray=cv2.Canny(image_gray,80,350)
plt.imshow(gray,cmap='gray')
```

```python
conrours , hierachy =cv2.findContours(gray,cv2.RETR_LIST , cv2.CHAIN_APPROX_SIMPLE) 

## Mode --- Hierachy relationship between contours 
## Method -- countours points are stored 

countour_image=image.copy()


cv2.drawContours(countour_image,conrours,-1,(0,255,0),5)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.title("original image")

plt.subplot(1,3,2)
plt.imshow(gray,cmap='gray')
plt.title("Mask image")

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(countour_image,cv2.COLOR_BGR2RGB))
plt.title("countour image")

plt.show()
```

## Using Threshold

```python
mask,binary =cv2.threshold(image_gray,200,255,cv2.THRESH_BINARY)
plt.imshow(binary,cmap='gray')

conrours , hierachy =cv2.findContours(binary,cv2.RETR_LIST , cv2.CHAIN_APPROX_SIMPLE) 

## Mode --- Hierachy relationship between contours 
## Method -- countours points are stored 

countour_image=image.copy()


cv2.drawContours(countour_image,conrours,-1,(0,255,0),5)


plt.figure(figsize=(10,5))

plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.title("original image")

plt.subplot(1,3,2)
plt.imshow(binary,cmap='gray')
plt.title("Mask image")

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(countour_image,cv2.COLOR_BGR2RGB))
plt.title("countour image")

plt.show()
```

## Target spasific Color

```python
lower_bound=np.array([120,100,0],dtype=np.uint8)
Upper_bound=np.array([160,185,255],dtype=np.uint8)
image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image_hsv=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
mask=cv2.inRange(image_hsv,lower_bound,Upper_bound) ### To defiend the color of an object 

plt.imshow(mask,cmap='gray')

conrours , hierachy =cv2.findContours(mask,cv2.RETR_LIST , cv2.CHAIN_APPROX_SIMPLE) 

## Mode --- Hierachy relationship between contours 
## Method -- countours points are stored 

countour_image=image.copy()


cv2.drawContours(countour_image,conrours,-1,(0,255,0),5) ## 5 value is the thikness of the line boundaries


plt.figure(figsize=(10,5))

plt.subplot(1,3,1)
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.title("original image")

plt.subplot(1,3,2)
plt.imshow(binary,cmap='gray')
plt.title("Mask image")

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(countour_image,cv2.COLOR_BGR2RGB))
plt.title("countour image")

plt.show()
```

## Rectangle Shape 

```python
lower_bound=np.array([200],dtype=np.uint8)
Upper_bound=np.array([255],dtype=np.uint8)

image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image_hsv=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
image_gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

mask=cv2.inRange(image_gray,lower_bound,Upper_bound) ### To defiend the color of an object 

plt.imshow(mask,cmap='gray')

conrours , hierachy =cv2.findContours(mask,cv2.RETR_LIST , cv2.CHAIN_APPROX_SIMPLE) 

## Mode --- Hierachy relationship between contours 
## Method -- countours points are stored 

countour_image=image.copy()


cv2.drawContours(countour_image,conrours,-1,(0,255,0),-1) ## 5 or -1  value is the thikness of the line boundaries :   -1 : means fill all the shape not only the boundaries 


plt.figure(figsize=(10,5))

plt.subplot(1,3,1)
plt.imshow(image_gray,cmap='gray')
plt.title("original image")

plt.subplot(1,3,2)
plt.imshow(mask,cmap='gray')
plt.title("Mask image")

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(countour_image,cv2.COLOR_BGR2RGB))
plt.title("countour image")

plt.show()
```

```python
lower_bound=np.array([200],dtype=np.uint8)
Upper_bound=np.array([255],dtype=np.uint8)

image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
image_hsv=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
image_gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

mask=cv2.inRange(image_gray,lower_bound,Upper_bound) ### To defiend the color of an object 

plt.imshow(mask,cmap='gray')

conrours , hierachy =cv2.findContours(mask,cv2.RETR_LIST , cv2.CHAIN_APPROX_SIMPLE) 

sorted_countour=sorted(conrours,key=cv2.contourArea,reverse=True)


## Mode --- Hierachy relationship between contours 
## Method -- countours points are stored 

print(len(sorted_countour))


conrours=sorted_countour[1:10]
```

```python
countour_image=image.copy()

for countour in conrours : 
    x,y,w,h=cv2.boundingRect(countour)
    cv2.rectangle(countour_image,(x,y),(x+w,y+h),(0,0,255),2)


plt.figure(figsize=(10,5))

plt.subplot(1,3,1)
plt.imshow(image_gray,cmap='gray')
plt.title("original image")

plt.subplot(1,3,2)
plt.imshow(mask,cmap='gray')
plt.title("Mask image")

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(countour_image,cv2.COLOR_BGR2RGB))
plt.title("countour image")

plt.show()
```
